```python
import numpy as np
import math

##Initializing the variables (will rename them to a0, sg0 etc. for clarity)
sg= np.ones(2)
a = np.ones(2)
b =  [2,3]
c=0
d = [4,5]
e=6
f= [7,8]
w = np.zeros((2,2))
 
x = np.array([[1,1],[1,-1],[-1,1],[1,1]])

# params are sg,a,b,c,d,e,f,w
def A(sg1,sg2):
  return 3
def px(sg,a,b,c,d,e,f,w): # test P(X)
  sg = np .array(sg)
  if (sg == [1,1]).all():
    return 0.25
  else:
    return 1
    ### some fumction of the first two  
print(px(sg,a,b,c,d,e,f,w))
print(x.shape[0]) 
def sx(sg,a,b,c,d,e,f,w):  ## S(x), i,j runs from 1 to 2
  term1= c+np.einsum('i,i->', d, sg) 
  term2 =(e+np.einsum('i,i', f, sg))*1j
  return math.tanh(term1+term2)
####
def Bxa(sg,a,b,c,d,e,f,w): #B(x;a_i)
  return 0.5*sg
def Bxb (sg,a,b,c,d,e,f,w):
  return 0.5*np.array(list(map(math.tanh, b+np.einsum('ij,i->j', w, sg))))
def Bxc (sg,a,b,c,d,e,f,w):
  return (1/( sx(sg,a,b,c,d,e,f,w))- sx(sg,a,b,c,d,e,f,w))
def Bxd (sg,a,b,c,d,e,f,w):
  return  (1/( sx(sg,a,b,c,d,e,f,w))- sx(sg,a,b,c,d,e,f,w))*sg
def Bxe (sg,a,b,c,d,e,f,w):
  return (1/( sx(sg,a,b,c,d,e,f,w))- sx(sg,a,b,c,d,e,f,w))*1j

def Bxf (sg,a,b,c,d,e,f,w):
  return (1/( sx(sg,a,b,c,d,e,f,w))- sx(sg,a,b,c,d,e,f,w))*1j*sg
def Bxw (sg,a,b,c,d,e,f,w):
  term1 = 0.5*np.array(list(map(math.tanh, b+np.einsum('ij,i->j', w, sg))))
  term2 = sg
  return np.einsum('j,i->ij',term1,term2)
######
def Avx(Bxi,sg,a,b,c,d,e,f,w): #<B(x;p_k)>_x  this doesn't depend on sg actually
  denom=0
  for i in range(x.shape[0]):
    denom+= abs(math.sqrt(px(x[i],a,b,c,d,e,f,w))*sx(x[i],a,b,c,d,e,f,w))**2
####
  
  if Bxi == Bxc or Bxi == Bxe :
   numer=0
   for i in range(x.shape[0]):
      t1= np.conj(math.sqrt(px(x[i],a,b,c,d,e,f,w))*sx(x[i],a,b,c,d,e,f,w))
      t2= Bxi(x[i],a,b,c,d,e,f,w)
      t3= math.sqrt(px(x[i],a,b,c,d,e,f,w))*sx(x[i],a,b,c,d,e,f,w)
      numer+=t1*t2*t3
   return (numer)/(denom)
  elif Bxi == Bxw:
     numer=np.zeros((w.shape[0],w.shape[0]))
     for l in range(a.shape[0]):
       for k in range(a.shape[0]):
         numerkl=0
         for i in range(x.shape[0]):
            t1= np.conj(math.sqrt(px(x[i],a,b,c,d,e,f,w))*sx(x[i],a,b,c,d,e,f,w))
            t2= Bxi(x[i],a,b,c,d,e,f,w)[k][l]
            t3= math.sqrt(px(x[i],a,b,c,d,e,f,w))*sx(x[i],a,b,c,d,e,f,w)
            numerkl+=t1*t2*t3
         numer[k][l]=numerkl
     return (numer)/(denom)
  else:
     numer=np.zeros(a.shape[0])
     for k in range(a.shape[0]):
       numerk=0
       for i in range(x.shape[0]):
         t1= np.conj(math.sqrt(px(x[i],a,b,c,d,e,f,w))*sx(x[i],a,b,c,d,e,f,w))
         t2= Bxi(x[i],a,b,c,d,e,f,w)[k]
         t3= math.sqrt(px(x[i],a,b,c,d,e,f,w))*sx(x[i],a,b,c,d,e,f,w)
         numerk+=t1*t2*t3
       numer[k]=numerk
     return (numer)/(denom)    
#################




def Avxx(Bxi,sg,a,b,c,d,e,f,w): #<B(x;p_k)>_x',x  #Done
  denom=0
  for i in range(x.shape[0]):
    denom+= abs(math.sqrt(px(x[i],a,b,c,d,e,f,w))*sx(x[i],a,b,c,d,e,f,w))**2
####
  
  if Bxi == Bxc or Bxi == Bxe :
   numer=0
   for i in range(x.shape[0]):
    for j in range(x.shape[0]):
      t1= np.conj(math.sqrt(px(x[i],a,b,c,d,e,f,w))*sx(x[i],a,b,c,d,e,f,w))
      t2= A(x[i],x[j])*Bxi(x[j],a,b,c,d,e,f,w)
      t3= math.sqrt(px(x[j],a,b,c,d,e,f,w))*sx(x[j],a,b,c,d,e,f,w)
      numer+=t1*t2*t3
   return (numer)/(denom)
  elif Bxi == Bxw:
     numer=np.zeros((w.shape[0],w.shape[0]))
     for l in range(a.shape[0]):
       for k in range(a.shape[0]):
         numerkl=0
         for i in range(x.shape[0]):
           for j in range(x.shape[0]):
            t1= np.conj(math.sqrt(px(x[i],a,b,c,d,e,f,w))*sx(x[i],a,b,c,d,e,f,w))
            t2= A(x[i],x[j])*Bxi(x[j],a,b,c,d,e,f,w)[k][l]
            t3= math.sqrt(px(x[j],a,b,c,d,e,f,w))*sx(x[j],a,b,c,d,e,f,w)
            numerkl+=t1*t2*t3
         numer[k][l]=numerkl

     
     return (numer)/(denom)
  else:
     numer=np.zeros(a.shape[0])
     for k in range(a.shape[0]):
       numerk=0
       for i in range(x.shape[0]):
         for j in range(x.shape[0]):
          t1= np.conj(math.sqrt(px(x[i],a,b,c,d,e,f,w))*sx(x[i],a,b,c,d,e,f,w))
          t2= A(x[i],x[j])*Bxi(x[j],a,b,c,d,e,f,w)[k]
          t3= math.sqrt(px(x[j],a,b,c,d,e,f,w))*sx(x[j],a,b,c,d,e,f,w)
          numerk+=t1*t2*t3
       numer[k]=numerk
     return (numer)/(denom)    

 



def AvA (sg,a,b,c,d,e,f,w):  #<A> doesn't depend on sg actually #Done
  numer=0
  for i in range(x.shape[0]):
    for j in range(x.shape[0]):
      t1= np.conj(math.sqrt(px(x[i],a,b,c,d,e,f,w))*sx(x[i],a,b,c,d,e,f,w))
      t2= A(x[i],x[j])
      t3= math.sqrt(px(x[j],a,b,c,d,e,f,w))*sx(x[j],a,b,c,d,e,f,w)
      numer+=t1*t2*t3
  denom=0
  for i in range(x.shape[0]):
    denom+= abs(math.sqrt(px(x[i],a,b,c,d,e,f,w))*sx(x[i],a,b,c,d,e,f,w))**2
  return (numer)/(denom)




 

def gd(grad, sg,a,b,c,d,e,f,w, n_epochs=3, eta=0.01):  #Gradient descent
    #We can here use Adam instead of this very simple descent
    
    term1=np.append(np.concatenate((sg,a,b)),c) 
    term2 = np.append(term1,d)
    term3=np.append(term2,e)
    term4 = np.append(term3,f)
    params = np.append(term4,w)
    param_traj=np.zeros([n_epochs+1,params.shape[0]])
    param_traj[0,]=params
    v=0;
    for j in range(n_epochs):
         v=eta*(np.array(grad(params)))
         params=params-v
         param_traj[j+1,]=params

    return  param_traj

def partialAx(Bxi,sg,a,b,c,d,e,f,w): #partial<A>/partial x_k
  term1 = Avxx(Bxi,sg,a,b,c,d,e,f,w) 
  if Bxi == Bxc or Bxi == Bxe:
    term2 = term1+np.conjugate(term1)
  else:
    term2 = term1 + np.array(list(map(np.conj, term1)))
  term3 = Avx(Bxi,sg,a,b,c,d,e,f,w) 
  if Bxi == Bxc or Bxi == Bxe:
    term4 = AvA(sg,a,b,c,d,e,f,w)*(term3+ np.conjugate(term3))
  else:
    term4 = 0#AvA(sg,a,b,c,d,e,f,w)*(term3+ np.array(list(map(np.conj, term3))))
  return term2+term4
def grad(params): # vectorial gradient
  sg = params[0:2]
  sg= np.array(sg)
  a  =  np.array(params[2:4])
  b  =  np.array(params[4:6])
  c  =  np.array(params[6])
  d  =  np.array(params[7:9])
  e  =  np.array(params[9])
  f  =  np.array(params[10:12])
  w  =  np.array(np.resize(params[12:16],(2,2))) #
 
  term1=np.append(np.concatenate((sg,partialAx(Bxa,sg,a,b,c,d,e,f,w),partialAx(Bxb,sg,a,b,c,d,e,f,w))),partialAx(Bxc,sg,a,b,c,d,e,f,w)) 
  term2 = np.append(term1,partialAx(Bxd,sg,a,b,c,d,e,f,w))
  term3= np.append(term2,partialAx(Bxe,sg,a,b,c,d,e,f,w))
  term4 = np.append(term3,partialAx(Bxf,sg,a,b,c,d,e,f,w))
  params = np.append(term4,partialAx(Bxw,sg,a,b,c,d,e,f,w))
  return np.around(params,5).real

print(Avxx(Bxw,sg,a,b,c,d,e,f,w))
print( gd(grad, sg,a,b,c,d,e,f,w, n_epochs=3, eta=0.01))
  
############################## END of NEW CODE
```
